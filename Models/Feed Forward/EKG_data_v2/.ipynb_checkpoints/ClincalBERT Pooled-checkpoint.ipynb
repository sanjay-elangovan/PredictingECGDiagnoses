## Imports

```python
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras import regularizers
import pickle
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
import scipy.sparse 
from matplotlib import pyplot as plt
```

## Loading Data

### tf-idf Embeddings

```python
embeddings = np.load('/home/sanjaycollege15/PredictingDiagnoses/Data/ekg_denoised_ClinicalBERT_embeddings_30k.npy')
labels = np.load('/home/sanjaycollege15/PredictingDiagnoses/Data/ekg_denoised_ClinicalBERT_labels_30k.npy')
```

```python
train_X, test_X, train_y, test_y = train_test_split(embeddings, labels, test_size=0.20, random_state=42, stratify=labels)
del embeddings
del labels
```

```python
train_X.shape
```

```python
test_X.shape
```

## Define Keras Model

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Input(shape=(264,768)))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(700, activation='relu', kernel_regularizer=regularizers.l1_l2(l1=1e-5, l2=1e-4)))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(200, activation='relu', kernel_regularizer=regularizers.l1_l2(l1=1e-5, l2=1e-4)))
model.add(tf.keras.layers.Dense(4, activation='softmax'))

model.build()
model.compile(loss='SparseCategoricalCrossentropy', 
              optimizer='adam', 
              metrics=['sparse_categorical_accuracy'])
```

```python
tf.keras.utils.plot_model(model, show_shapes = True)
```

```python
model.fit(train_X, train_y, epochs=10, batch_size=512, validation_split=0.2)
```

```python
model.evaluate(test_X, test_y)
```

Some learning, about the same as for the RNN models actually (31%).
